```python
import sympy as smp
import numpy as np
import time
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
#def timer(f):
#  def wrapper(*args, **kwargs):
#    t0 = time.time()
#    result = f(*args, **kwargs)
#    t1 = time.time()
#    return [result, 1000*(t1-t0)]
#  return wrapper


def harmonic_f(x):
  return np.sin(2*x*np.pi) * 5

def log_f(x):
  return 2*(np.log( x*11 ) + 3)

def polynomial_f(x):
  return x**2 -3


x = np.linspace(1,5, 100)
```

```python
print(f"Lichoběžníkové pravidlo.\nPolynomial: {integrate.trapezoid(polynomial_f(x), x = x)}")
print(f"Romberg.\nPolynomial: {integrate.romberg(polynomial_f, 1, 5)}")
print(f"Simpson.\nPolynomial: {integrate.simpson(polynomial_f(x), x = x)}")

plt.plot(x, polynomial_f(x), label="Polynomial", color="green")
plt.legend()
plt.show()
```

```python
print(f"Lichoběžníkové pravidlo.\nLogarithmic: {integrate.trapezoid(log_f(x), x = x)}")
print(f"Romberg.\nLogarithmic: {integrate.romberg(log_f, 1, 5)}")
print(f"Simpson.\nLogarithmic: {integrate.simpson(log_f(x), x = x)}")

plt.plot(x, log_f(x), label="Logarithmic", color="red")
plt.legend()
plt.show()
```

```python
print(f"Lichoběžníkové pravidlo.\nHarmonic: {integrate.trapezoid(harmonic_f(x), x = x)}")
print(f"Romberg.\nHarmonic: {integrate.romberg(harmonic_f, 1, 5)}")
print(f"Simpson.\nHarmonic: {integrate.simpson(harmonic_f(x), x = x)}")

plt.plot(x, harmonic_f(x), label="Harmonic", color="yellow")
plt.legend()
plt.show()
```
