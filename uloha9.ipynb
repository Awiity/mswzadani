```python
import sympy as smp
import numpy as np
import time
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
#def timer(f):
#  def wrapper(*args, **kwargs):
#    t0 = time.time()
#    result = f(*args, **kwargs)
#    t1 = time.time()
#    return [result, 1000*(t1-t0)]
#  return wrapper


def harmonic_f(x):
  return np.sin(2*x*np.pi) * 5

def log_f(x):
  return 2*(np.log( x*11 ) + 3)

def polynomial_f(x):
  return x**2 -3


x = np.linspace(1,5, 50)

plt.plot(x, polynomial_f(x), label="Polynomial")
plt.plot(x, harmonic_f(x), label="Harmonic")
plt.plot(x, log_f(x), label="Logarithmic")
plt.legend()
plt.show()

print("Lichoběžníkové pravidlo.")
print(f"Polynomial: {integrate.trapezoid(polynomial_f(x), x = x)}")
print(f"Logarithmic: {integrate.trapezoid(log_f(x), x = x)}")
print(f"Harmonic: {integrate.trapezoid(harmonic_f(x), x = x)}\n")

print("Romberg")
print(f"Polynomial: {integrate.romberg(polynomial_f, 1, 5)}")
print(f"Logarithmic: {integrate.romberg(log_f, 1, 5)}")
print(f"Harmonic: {integrate.romberg(harmonic_f, 1, 5)}\n")

print("Simpson")
print(f"Polynomial: {integrate.simpson(polynomial_f(x), x = x)}")
print(f"Logarithmic: {integrate.simpson(log_f(x), x = x)}")
print(f"Harmonic: {integrate.simpson(harmonic_f(x), x = x)}\n")
```
