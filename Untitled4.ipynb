```python
import scipy
import matplotlib.pyplot as plt
import numpy as np
import sympy
import time
import pandas as pd
import seaborn as sns


df = pd.read_csv("supermarket_sales.csv")
print(f"shape:{df.shape},\ncols:{df.columns}")
df['Date'] = pd.to_datetime(df['Date'])
print(df['Date'][0])
df.set_index('Date', inplace=True)

mean_unit_price = df['Unit price'].astype("float").mean(axis=0)
df["Unit price"].replace(np.nan, mean_unit_price)
df.dropna()
```

```python
t1 = time.time()
plt.figure(figsize=(10,5), dpi=100)

plt.title("Correlation between price of goods and gross income of the costumer")
sns.regplot(x="Unit price", y="gross income", data=df, color='y')
plt.xlabel("Unit Price")
plt.ylabel("Gross Income")
plt.show()

t2 = time.time()
print(f"TIME: {(t2 - t1)*1000}ms")
```

```python
t1 = time.time()
plt.figure(figsize=(12,6),dpi=100)
sns.displot(df['Rating'])
plt.axvline(np.mean(df['Rating']), color='r', label="Average Rating Score")
plt.legend()
plt.show()

t2 = time.time()
print(f"TIME:{(t2-t1)*1000}ms")
```

```python
t1=time.time()
my_palette = sns.color_palette("muted")
plt.figure(figsize=(10,5), dpi=100)

sns.countplot(x=df['Branch'], hue=df['Branch'], palette = my_palette)
plt.xlabel('Branches')
plt.ylabel('Count of operations')
plt.title("sss")

print(df.groupby('Branch')['Total'].count())

plt.show()
t2=time.time()
print(f"TIME:{(t2-t1)*1000}ms")
```

```python
t1=time.time()

plt.figure(figsize=(10,5), dpi=100)

sns.countplot(x=df['Payment'], hue=df['City'], palette = "muted")
plt.xlabel("Payment method used")
plt.ylabel("Count of cases")

print(df.groupby('Payment')['Total'].count())

plt.show()
t2=time.time()
print(f"TIME:{(t2-t1)*1000}ms")
```
